```python
ddf = df.groupby('date_time')
```

```python
ddf.head(2)
ddf = ddf.apply(lambda x: x.assign(items=[{'name': row['name'], 'quantity': row['quantity']} for _, row in x.iterrows()]))
```

```python
ddf.reset_index(drop=True, inplace=True)
ddf['concatenated_items'] = ddf.groupby('date_time')['items'].transform(lambda x: ', '.join(map(str, x)))
```
